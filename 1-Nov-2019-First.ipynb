```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path ="movie_metadata.csv"
data = pd.read_csv(path)
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
numerical_cols = data.columns\
                [data.dtypes != object]
```

```python
for i in range(len(numerical_cols)):
    data[numerical_cols[i]].fillna(\
        data[numerical_cols[i]].mean(),inplace=True)
```

```python
data[numerical_cols].isnull().sum()
```

```python
target = data['imdb_score']
num_features = data[numerical_cols].drop('imdb_score',axis=1)
num_features.head()
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test = \
            train_test_split(num_features,target,random_state=23)
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr = LinearRegression()
lr.fit(x_train,y_train)
mean_squared_error(y_test,lr.predict(x_test))
```

```python
check = pd.DataFrame(lr.predict(x_test),\
        index = x_test.index,columns=['pred']) 
check[:5]
```

```python
result = pd.concat([new,check],axis=1)
result.head()
```
